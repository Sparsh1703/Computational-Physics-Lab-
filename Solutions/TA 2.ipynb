```python
import numpy as np
import matplotlib.pyplot as plt
```

## Question 1

```python
h = 0.1
x = np.arange(1,3+h,h)
x_0 = 1
y_0 = 5

def F(x,y):
    return x*y

def Euler(x_0, y_0, h, x):
    y_euler = [y_0]
    
    for i in x:
        y = y_0 + h * F(i, y_0)
        y_euler.append(y)
        y_0 = y
        
    return y_euler
```

```python
y_euler = Euler(x_0, y_0, h, x)
y_euler = y_euler[:-1]
```

```python
plt.plot(x,y_euler)
```

```python
def Improved_Euler(x_0, y_0, h, x):
    y_Improved_euler = [y_0]
    
    for i in x:
        y = y_0 + h * F(i + h/2, y_0)
        y_Improved_euler.append(y)
        y_0 = y
        
    return y_Improved_euler    
```

```python
y_Improved_euler = Improved_Euler(x_0, y_0, h, x)
y_Improved_euler = y_Improved_euler[:-1]
```

```python
plt.plot(x,y_Improved_euler)
```

```python
def RK2(x_0,y_0,h,x):
    y_RK2 = [y_0]
    for i in x:
        k1 = h * F(i,y_0)
        k2 = h * F(i + h/2, y_0 + k1/2)
        y = y_0 + k2
        
        y_RK2.append(y)
        y_0 = y
        
    return y_RK2
```

```python
y_RK2 = RK2(x_0, y_0, h, x)
y_RK2 = y_RK2[:-1]
```

```python
plt.plot(x,y_RK2)
```

```python
plt.plot(x,y_euler, label='euler')
plt.plot(x,y_Improved_euler, label='Improved_euler')
plt.plot(x,y_RK2, label='RK2')
plt.legend()
plt.show()
```

## Question 2

```python
y_0 = 0.5
t_0 = 0
h = 0.1
t = np.arange(0,2+h,h)
```

```python
def F(t,y):
    return y - t**2 + 1

def RK2(t_0,y_0,h,t):
    y_RK2 = [y_0]
    for i in t:
        k1 = h * F(i,y_0)
        k2 = h * F(i + h/2, y_0 + k1/2)
        y = y_0 + k2
        
        y_RK2.append(y)
        y_0 = y
        
    return y_RK2
```

```python
y_RK2 = RK2(t_0, y_0, h, t)
y_RK2 = y_RK2[:-1]
```

```python
plt.plot(t,y_RK2)
```

```python
y_analytical = t**2 + 2*t + 1 - 0.5*np.exp(t)
plt.plot(t,y_analytical)
```

```python
plt.plot(t,y_RK2, label='RK2')
plt.plot(t,y_analytical, label='analytical')
plt.legend()
plt.show()
```

```python
error = y_analytical - y_RK2
plt.plot(t,error)
plt.xlabel('t')
plt.ylabel('Error')
```

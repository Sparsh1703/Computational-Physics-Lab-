```python
import numpy as np
import matplotlib.pyplot as plt
```

## Question 1

```python
v=1.0
h=0.01
t=h/2
```

```python
#Central Difference method
u = np.zeros((200,100))
u[0]=1
u[1][0]=1

for n in range (0,199):
    for j in range (0,100):
        if j==0 or j==99:
            u[n+1][j]=u[n][j]
        else:
            u[n+1][j]= u[n][j] - v*t/(2*h)*(u[n][j+1]-u[n][j-1])

            
fig = plt.figure(figsize=(10,10))            
ax = plt.axes(projection='3d')

x = np.linspace(0,1,100)
y= np.linspace(0,1,200)
ax.set_xlabel('$x$')
ax.set_ylabel('$t$')
ax.set_zlabel('$u$')
X,Y = np.meshgrid(x,y)

ax.plot_surface(X,Y,u)
```

```python
#Upwind Method
u = np.zeros((200,100))
u[0]=1

print(u)
for n in range (0,199):
    for j in range (0,100):
        if j==0:
            u[n+1][j]=u[n][j]
        else:
            u[n+1][j]= u[n][j] - v*t/(h)*(u[n][j]-u[n][j-1])

fig = plt.figure(figsize=(10,10))            
ax = plt.axes(projection='3d')

x = np.linspace(0,1,100)
y= np.linspace(0,1,200)
ax.set_title('Upwind method')
ax.set_xlabel('$x$')
ax.set_ylabel('$t$')
ax.set_zlabel('$u$')
ax.dimensions?
X,Y = np.meshgrid(x,y)

ax.plot_surface(X,Y,u)
```

## Question 2

```python
t=0.0025
h=0.1

u = np.zeros((10000,10))

for i in range(0,10):
    u[0][i]= np.sin(3*np.pi/2*h*i)

for n in range (0,9999):
    for j in range (0,10):
        if j==0:
            u[n+1][j]=0
        elif j==9:
            u[n+1][j]=u[n+1][j-1]
        else:
            u[n+1][j]= u[n][j] + t/(h**2)*(u[n][j+1]-2*u[n][j]+u[n][j-1])
                                           
x = np.arange(0,1,0.1)
t0 = np.arange(0,10000*0.0025,0.0025)

X,T = np.meshgrid(x,t0)

fig = plt.figure(figsize=(10,10))            
ax = plt.axes(projection='3d')


ax.set_xlabel('$x$')
ax.set_ylabel('$t$')
ax.set_zlabel('$u$')
ax.set_title('Approximated Function')
#X.shape
ax.plot_surface(X,T,u)
```

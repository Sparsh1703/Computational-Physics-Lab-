```python
import numpy as np
import matplotlib.pyplot as plt
```

## Question 1

```python
h = 0.1
x = np.arange(0,6+h,h)
x_0 = 0
y_0 = 5
```

```python
def F(x,y):
    return (x + y) * np.sin(x)

def RK2(x_0,y_0,h,x):
    y_RK2 = [y_0]
    for i in x:
        k1 = h * F(i,y_0)
        k2 = h * F(i + h/2, y_0 + k1/2)
        y = y_0 + k2
        
        y_RK2.append(y)
        y_0 = y
        
    return y_RK2
```

```python
y_RK2 = RK2(x_0,y_0,h,x)
y_RK2 = y_RK2[:-1]
```

```python
plt.plot(x,y_RK2)
```

```python
def RK4(x_0,y_0,h,x):
    y_RK4 = [y_0]
    for i in x:
        k1 = h * F(i,y_0)
        k2 = h * F(i + h/2, y_0 + k1/2)
        k3 = h * F(i + h/2, y_0 + k2/2)
        k4 = h * F(i + h, y_0 + k3)
        y = y_0 + (k1/2 + k2 + k3 + k4/2)/3
        
        y_RK4.append(y)
        y_0 = y
        
    return y_RK4
```

```python
y_RK4 = RK4(x_0,y_0,h,x)
y_RK4 = y_RK4[:-1]
```

```python
plt.plot(x,y_RK4)
```

```python
plt.plot(x,y_RK2, label='RK2')
plt.plot(x,y_RK4, label='RK4')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Question 2

```python
h = 0.1
x = np.arange(0, 3+h, h)
x_0 = 0
y_0 = 0
v_0 = 5               # v = dy/dx
```

```python
# Euler's Method
y_euler = [y_0]
v_euler = [v_0]

for i in x:
    v = v_0 + h * (-5*v_0 - 6*y_0 + 10*np.sin(i))
    y = y_0 + h*v_0
    
    v_euler.append(v)
    v_0 = v
    y_euler.append(y)
    y_0 = y
```

```python
y_euler = y_euler[:-1]
```

```python
plt.plot(x,y_euler)
```

```python
# RK4 Method
v_RK4 = [v_0]
y_RK4 = [y_0]

for i in x:
    k1 = h * (10*np.sin(i) - 6*y_0 - 5*v_0)
    k2 = h * (10*np.sin(i + h/2) - 6*y_0 - 5*(v_0 + k1/2))
    k3 = h * (10*np.sin(i + h/2) - 6*y_0 - 5*(v_0 + k2/2))
    k4 = h * (10*np.sin(i + h) - 6*y_0 - 5*(v_0 + k3))
    
    p1 = h * v_0
    p2 = h * (v_0)
    p3 = h * (v_0)
    p4 = h * (v_0)
    
    v = v_0 + (k1/2 + k2 + k3 + k4/2)/3
    y = y_0 + (p1/2 + p2 + p3 + p4/2)/3
    
    v_RK4.append(v)
    y_RK4.append(y)
    v_0 = v
    y_0 = y
```

```python
y_RK4 = y_RK4[:-1]
```

```python
plt.plot(x,y_RK4)
```

```python
y_analytical = np.sin(x) - np.cos(x) - 6*np.exp(-3*x) + 7*np.exp(-2*x)
plt.plot(x,y_analytical)
```

```python
plt.plot(x,y_analytical,label = 'analytical')
plt.plot(x,y_euler, label = 'euler')
plt.plot(x,y_RK4, label = 'RK4')
plt.legend()
plt.show()
```
